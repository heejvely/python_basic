```python
# Q. 네이버 사이트에서 strong 태그를 이용하여 모든 텍스트를 가져와서 
# 출력하세요.
elems = driver.find_elements_by_tag_name('strong')
for e in elems:
    print(e.text)
```

```python
# Q. 네이버 사이트에서 id를 이용하여 다음을 출력하세요.
# 뉴스스탠드
# 구독한 언론사 전체언론사
# 리스트형
# 썸네일형
# 설정
elem = driver.find_element_by_id('NM_NEWSSTAND_TITLE')
print(elem.text)
```
