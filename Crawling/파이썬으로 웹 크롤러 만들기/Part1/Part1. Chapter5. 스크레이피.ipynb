```python
import scrapy

class ArticleSpider(scrapy.Spider):
    name = 'article'
    
    def start_requests(self):
        urls = [
            'http://en.wikipedia.org/wiki/Python_',
            '%28programming_language%29',
            'https://en.wikipedia.org/wiki/Functional_programming',
            'https://en.wikipedia.org/wiki/Monty_Python'
        ]
        return [scrapy.Request(url = url, callback = self.parse) for url in urls]
    
    def parse(self, response):
        url = response.url
        title = response.css('h1::text').extract_first()
        print(f'URL is: {url}')
        print(f'Title is: {title}')
```

1. link_extractor
: 필수 매개변수인 LinkExtractor 객체입니다. 필수 매개변수는 이 객체뿐입니다.
<br/>

2. callback
: 페이지 내용을 구문 분석하는 함수입니다.
<br/>

3. cb_kwargs
: 콜백 함수에 전달할 매개변수 딕셔너리입니다. 이 딕셔너리는 {arg_name1: arg_value1, arg_name2: arg_value2} 형식으로 되어있으며, 약간 다른 작업에 동일한 파서 함수를 재사용하기 편리합니다.
<br/>

4. follow
: 현재 페이지의 링크를 향후 크롤링에 포함할지 여부를 나타냅니다. 콜백 함수가 제공되지 않으면 기본값은 True입니다.(현재 페이지에서 아무것도 하지 않았다면 최소한 크롤링을 계속하긴 해야하니까요). 콜백함수가 제공되면 기본값은 False입니다.
<br/>

5. allow
: 제공된 정규표현식과 일치하는 링크를 모두 허용합니다.
<br/>

6. deny
: 제공된 정규 표현식과 일치하는 링크를 모두 거부합니다.
