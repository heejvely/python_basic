```python
# 람다함수와 적합한 내장함수를 사용하여 아래와 같이 출력하세요.
#[1,2.3,4,5] -> [1,4,9,16,25]
#[1,2,3,4,5] -> [1,2]

ab = [1,2,3,4,5]

print(list(map(lambda i : i*i, ab))) # 모든 요소 for 문을 쓰지 않고 한 번에 변환
print(list(filter(lambda i : i<3, ab))) # 여러 개의 데이터를 일부만 추려낼 때 사용
```

```python
# [1,-3,2,0,-5,6]에서 양수만 필터링해서 출력하세요.(사용자 함수 이용)

a = [1,-3,2,0,-5,6]

def filt(x):
  plus = []
  for i in x:
    if i > 0:
      plus.append(i)
    else:
      pass
  return plus

filt(a)
```

```python
# 내장함수를 이용하여 동일한 개수로 이루어진 자료형을 화살표 좌흑에서 우측으로 변환하세요.
# [1,2,3],[4,5,6],[7,8,9] -> [(1,4,7),(2,5,8),(3,6,9)]
# ('abc','def') -> [('a','d'),('b','e'),('c','f')]

print(list(zip([1,2,3],[4,5,6],[7,8,9]))) # 각 원소를 튜플의 형태로 인덱스 번호에 맞춰 묶어줌.
print(list(zip('abc','def')))
```

```python
# 입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성한 후 1,2,3을 입력한 결과를 출력하세요.

def aver(*arg):
  sum = 0
  for i in arg:
    sum += i
  result = sum/len(arg)
  return result

aver(1,2,3)
```

```python
# strings를 다양한 문자가 포함된 순서로 정렬하세요.

strings = ['a','ca','bar','aaa','ababa']

strings.sort(key=lambda x : len(set(x)), reverse=True)
print(strings)
```

```python
# Q. strings를 다양한 문자가 포함된 순서로 정렬하세요.
strings = ['foo', 'card', 'bar', 'aaaa', 'abab']

def rev(x):
  return len(set(x))

strings.sort(key=rev,reverse=True)
print(strings)
```

```python
# 튜플 형태인 strings를 다양한 문자가 포함된 순서로 정렬하세요.
# 튜플을 리스트로 변경해서 작성
strings1 = {'foo','card','bar','aaaa','abab'}

strings = list(strings1)
strings.sort(key = lambda x : len(set(x)),reverse=True)

print(strings)
```

```python
# 게시물의 총 건수와 한 페이지에 보여 줄 게시물 수를 입력시 총 페이지 수를 출력하는 프로그램
import math

def allp(a,b):
  if a > 0 and b > 0:
    re = math.ceil(a/b) # 실수를 올림하여 정수를 반환하는 함수 ,내림: floor
  return re

allp(80,12)
```

```python
def page(a,b):
  if a % b == 0:
    return a//b
  else:
    return a//b + 1

page(80,12)
```

```python
# Q. 랜덤하게 1000명의 이름, 키와 몸무게 만들기

import random

hanguls = list('가나다라마바사아자차카타파하')
with open('index.txt','w') as f:
  for i in range(1000):
    name = random.choice(hanguls) + random.choice(hanguls)
    height = random.randint(140,200)
    weight = random.randint(40,100)
    f.write(f'{name}, {height}, {weight}\n')
```

```python
# 1000명에 대하여 아래와 같이 출력하세요.

with open('index.txt','r') as f:
  for line in f:
    (name, height, weight) = line.strip().split(',')
    print('\n'.join([f"이름:{name}",f"키:{height}",f"몸무게:{weight}"]))
    print()
```

```python
# long_task 함수는 수행하는 데 5초의 시간이 걸리는 함수이다. 위 프로그램은 이 함수를 수행하는 프로그램이다.
# 이 프로그램은 5초가 5번 반복되니 총 25초의 시간이 걸린다.

import time
def long_task(): # 5초의 시간이 걸리는 함수
  for i in range(5):
    time.sleep(1) 1초 대기
    print(f'working: {i}\n')

print('start')

for i in range(5): # long_task를 5회 수행(총 25초)
  long_task()

print('end')
```

```python
# join으로 스레드가 종료할 때까지 기다린다.

import time
import threading # 병렬처리

def long_task1():
  for i in range(5):
    time.sleep(1)
    print(f'working: {i}\n')

print('start')

threads = []
for i in range(5):
  t = threading.Thread(target = long_task1)
  threads.append(t)
for t in threads:
  t.start()
for t in threads:
  t.join() # 부모 thread를 진행하지 않고 새로 생성한 thread의 종료를 기다려준다는 의미 (끝까지 수행하기 위함)

print('end')
```
